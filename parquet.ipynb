```python
import pyarrow.parquet as pq
```

```python
trips = pq.read_table("yellow_tripdata_2022-02.parquet", use_legacy_dataset=True)
```

```python
pq.write_table(trips, "yellow_tripdata_uncompressed.parquet", compression="none")
pq.write_table(trips, "yellow_tripdata_snappy.parquet", compression="snappy")
pq.write_table(trips, "yellow_tripdata_lz4.parquet", compression="lz4")
pq.write_table(trips, "yellow_tripdata_gzip.parquet", compression="gzip")
```
